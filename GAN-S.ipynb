# Generative Adversarial Network for Spectrogras

So, what's a Generative Adversarial Network?

*"Two neural networks contest with each other in a zero-sum game framework. The generative network generates candidates while the discriminative network evaluates them."*
                                                                        
                         - Wikipedia

And what's a spectrogram?

*"A spectrogram is a visual representation of the spectrum of frequencies of a signal as it varies with time"*
                                                        
                         - Wikipedia

```python
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy import signal
import librosa
```

```python
def plot_spectrogram(title, times, frequencies, spectrogram, width=16, height=5):
    plt.figure(figsize=(width, height))
    plt.pcolormesh(times, frequencies, spectrogram)
    plt.title(title)
    plt.ylabel('Frequency (Hz)')
    plt.xlabel('Time (sec)')
    plt.show()
    
def plot_spectrogram_zoom(title, times, frequencies, spectrogram, y_end, width=16, height=5):
    plt.figure(figsize=(width, height))
    plt.pcolormesh(times, frequencies, spectrogram)
    plt.ylim(0, y_end)
    plt.title(title+' Zoom')
    plt.ylabel('Frequency (Hz)')
    plt.xlabel('Time (sec)')
    plt.show()
    
def plot_spectrogram_comparison(title, times_1, frequencies_1, spectrogram_1,
                                times_2, frequencies_2, spectrogram_2,
                                y_end, width=16, height=5):
    plt.figure(figsize=(width, height))    
    plt.subplot(2, 1, 1)
    plt.title(title[0])
    plt.pcolormesh(times_1, frequencies_1, spectrogram_1)
    plt.ylim(0, y_end)
    plt.ylabel('Frequency (Hz)')
    plt.xlabel('Time (sec)')
    
    plt.subplot(2, 1, 2)
    plt.title(title[1])
    plt.pcolormesh(times_2, frequencies_2, spectrogram_2)
    plt.ylim(0, y_end)
    plt.ylabel('Frequency (Hz)')
    plt.xlabel('Time (sec)')
        
    plt.show()
```

```python
sample_rate, samples = wavfile.read('./samples/guitar_mono.wav')
acustic_frequencies, acustic_times, acustic_spectrogram = signal.spectrogram(samples, sample_rate)
```

```python
plot_spectrogram('Acustic Guitar', acustic_times, acustic_frequencies, acustic_spectrogram)
```

```python
plot_spectrogram_zoom('Acustic Guitar', acustic_times, acustic_frequencies, acustic_spectrogram, 4000, 16, 4)
```

```python
sample_rate, samples = wavfile.read('./samples/powermetal_mono.wav')
metal_frequencies, metal_times, metal_spectrogram = signal.spectrogram(samples, sample_rate)
```

```python
plot_spectrogram('Metal Guitar', metal_times, metal_frequencies, metal_spectrogram)
```

```python
plot_spectrogram_zoom('Metal Guitar', metal_times, metal_frequencies, metal_spectrogram, 4000, 16, 3)
```

```python
plot_spectrogram_comparison(['Metal Guitar','Acustic Guitar'],
                            metal_times, metal_frequencies, metal_spectrogram,
                            acustic_times, acustic_frequencies, acustic_spectrogram, 4000, 16, 10)
```
